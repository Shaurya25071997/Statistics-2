## Q5.When is it appropriate to refer to a skewed data distribution?

Skewed data distribution is the measure of the symmetry of the distribution It helps to understand where the data is lying and analyzing the outliers in a given data

It can be skewed in 3 formats:-

1. Positive skew or Right skew :- eg-> Wealth distribution, length of comment
2. Symmetry distribution or Normal distribution :- eg-> ages, weight etc
3. Negative skew or left skew :- eg -> life span of human being ages = 50-70
    
When these formats happens in data then this skewed form of data is given:-
Positive skew:- Mean>median>mode
symmetry :- Mean=Median=Mode
Negative skew:- Mode>Median>Mean
